# Quick Ride Rental(Mini project)

```python
class Vehicle:
    def __init__(self, vehicle_id, brand, model, price):
        self.vehicle_id = vehicle_id
        self.brand = brand
        self.model = model
        self.rental_price_per_day = price
        self.is_available = True

    def display_info(self):
        print(self.vehicle_id, self.brand, self.model,
              "Price:", self.rental_price_per_day,
              "Available:", self.is_available)

    def rent(self):
        if self.is_available:
            self.is_available = False
            print("Vehicle rented successfully")
        else:
            print("Vehicle not available")

    def return_vehicle(self):
        self.is_available = True

    def calculate_rental_cost(self, days):
        return self.rental_price_per_day * days

#Sub-class(Car)
class Car(Vehicle):
    def __init__(self, vehicle_id, brand, model, price, doors, fuel):
        super().__init__(vehicle_id, brand, model, price)
        self.num_doors = doors
        self.fuel_type = fuel

    def display_info(self):
        print("Car:", self.vehicle_id, self.brand, self.model,
              "Doors:", self.num_doors,
              "Fuel:", self.fuel_type,
              "Price:", self.rental_price_per_day,
              "Available:", self.is_available)

#Sub-class(Bike)
class Bike(Vehicle):
    def __init__(self, vehicle_id, brand, model, price, engine, bike_type):
        super().__init__(vehicle_id, brand, model, price)
        self.engine_capacity = engine
        self.bike_type = bike_type

    def display_info(self):
        print("Bike:", self.vehicle_id, self.brand, self.model,
              "Engine:", self.engine_capacity,
              "Type:", self.bike_type,
              "Price:", self.rental_price_per_day,
              "Available:", self.is_available)

#Customer class
class Customer:
    def __init__(self, customer_id, name):
        self.customer_id = customer_id
        self.name = name
        self.rented_vehicle = None

    def rent_vehicle(self, vehicle):
        self.rented_vehicle = vehicle

    def return_vehicle(self):
        self.rented_vehicle = None

    def view_rented_vehicle(self):
        if self.rented_vehicle:
            print("Rented Vehicle:", self.rented_vehicle.model)
        else:
            print("No vehicle rented")


#Rental sevvice
class RentalService:
    def __init__(self):
        self.vehicles = []
        self.customers = []

    def add_vehicle(self, vehicle):
        self.vehicles.append(vehicle)

    def register_customer(self, customer):
        self.customers.append(customer)

    def show_available_vehicles(self):
        for v in self.vehicles:
            if v.is_available:
                v.display_info()

    def find_vehicle(self, vehicle_id):
        for v in self.vehicles:
            if v.vehicle_id == vehicle_id:
                return v

    def find_customer(self, customer_id):
        for c in self.customers:
            if c.customer_id == customer_id:
                return c

    def rent_vehicle(self, customer_id, vehicle_id, days):
        customer = self.find_customer(customer_id)
        vehicle = self.find_vehicle(vehicle_id)

        if vehicle and vehicle.is_available:
            cost = vehicle.calculate_rental_cost(days)
            vehicle.rent()
            customer.rent_vehicle(vehicle)
            print("Total Cost:", cost, "PKR")
        else:
            print("Vehicle not available")

    def return_vehicle(self, customer_id):
        customer = self.find_customer(customer_id)

        if customer and customer.rented_vehicle:
            customer.rented_vehicle.return_vehicle()
            customer.return_vehicle()
            print("Vehicle returned successfully")

#Main part
service = RentalService()

bike1 = Bike("B101", "Honda", "CG125", 1000, "125cc", "Standard")
car1 = Car("C201", "Toyota", "Corolla", 4000, 4, "Petrol")
car2 = Car("C301", "Honda", "Civic", 6000, 4, "Petrol")

service.add_vehicle(bike1)
service.add_vehicle(car1)
service.add_vehicle(car2)

cust1 = Customer("U1", "Ali")
service.register_customer(cust1)


while True:
    print("Quick Ride Rental")
    print("\n1 Show Vehicles")
    print("2 Rent Vehicle")
    print("3 Return Vehicle")
    print("4 View Customer")
    print("5 Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        service.show_available_vehicles()

    elif choice == "2":
        vid = input("Enter vehicle id: ")
        days = int(input("Enter days: "))
        service.rent_vehicle("U1", vid, days)

    elif choice == "3":
        service.return_vehicle("U1")

    elif choice == "4":
        cust1.view_rented_vehicle()

    elif choice == "5":
        break
```

# Quick Ride Rental

```python
# Vehicle Class
class Vehicle:
    def __init__(self, vehicle_id, brand, model, rental_price_per_day):
        self.vehicle_id = vehicle_id
        self.brand = brand
        self.model = model
        self.rental_price_per_day = rental_price_per_day
        self.is_available = True
        
    def display_info(self):
        print(self.vehicle_id, self.brand, self.model,
              "Price:", self.rental_price_per_day,
              "Available:", self.is_available)
        
    def rent(self):
        if self.is_available:
            self.is_available = False
            return True
        return False
            
    def return_vehicle(self):
        self.is_available = True
            
    def calculate_rental_cost(self, days):
        return self.rental_price_per_day * days


# Car Class
class Car(Vehicle):
    def __init__(self, vehicle_id, brand, model, rental_price_per_day, doors, fuel):
        super().__init__(vehicle_id, brand, model, rental_price_per_day)
        self.num_doors = doors
        self.fuel_type = fuel
        
    def display_info(self):
        print("Car:", self.vehicle_id, self.brand, self.model,
              "Price:", self.rental_price_per_day,
              "Available:", self.is_available,
              "Doors:", self.num_doors,
              "Fuel:", self.fuel_type)


# Bike Class
class Bike(Vehicle):
    def __init__(self, vehicle_id, brand, model, rental_price_per_day, engine, bike_type):
        super().__init__(vehicle_id, brand, model, rental_price_per_day)
        self.engine_capacity = engine
        self.bike_type = bike_type
        
    def display_info(self):
        print("Bike:", self.vehicle_id, self.brand, self.model,
              "Price:", self.rental_price_per_day,
              "Available:", self.is_available,
              "Engine:", self.engine_capacity,
              "Type:", self.bike_type)


# Customer Class
class Customer:
    def __init__(self, customer_id, name):
        self.customer_id = customer_id
        self.name = name
        self.rented_vehicle = None
        
    def rent_vehicle(self, vehicle):
        if self.rented_vehicle:
            print("You already rented a vehicle")
        elif vehicle.is_available:
            if vehicle.rent():
                self.rented_vehicle = vehicle
                print(self.name, "rented", vehicle.model)
        else:
            print("Vehicle not available")
        
    def return_vehicle(self):
        if self.rented_vehicle:
            self.rented_vehicle.return_vehicle()
            print(self.name, "returned", self.rented_vehicle.model)
            self.rented_vehicle = None
        else:
            print("No vehicle to return")
        
    def view_rented_vehicle(self):
        if self.rented_vehicle:
            print("Rented Vehicle:", self.rented_vehicle.model)
        else:
            print("No vehicle rented")


# Rental Service Class
class RentalService:
    def __init__(self):
        self.vehicles = []
        self.customers = []
        
    def add_vehicle(self, vehicle):
        self.vehicles.append(vehicle)
        
    def register_customer(self, customer):
        self.customers.append(customer)
        
    def show_available_vehicles(self):
        print("\nAvailable Vehicles:")
        for v in self.vehicles:
            if v.is_available:
                v.display_info()
                
    def rent_vehicle(self, customer_id, vehicle_id, days):
        customer = None
        vehicle = None
        
        # Find customer
        for c in self.customers:
            if c.customer_id == customer_id:
                customer = c
                break
                
        # Find vehicle
        for v in self.vehicles:
            if v.vehicle_id == vehicle_id:
                vehicle = v
                break
                
        if customer and vehicle:
            if vehicle.is_available:
                cost = vehicle.calculate_rental_cost(days)
                customer.rent_vehicle(vehicle)
                print("Total Cost:", cost, "PKR")
            else:
                print("Vehicle not available")
        else:
            print("Invalid customer or vehicle ID")
                
    def return_vehicle(self, customer_id):
        for c in self.customers:
            if c.customer_id == customer_id:
                c.return_vehicle()
                return


# Main Program
service = RentalService()

# Vehicles
bike1 = Bike("B101", "Honda", "CG125", 1000, "150cc", "Standard")
car1 = Car("C102", "Toyota", "Corolla", 4000, 4, "Petrol")
car2 = Car("C103", "Honda", "Civic", 6000, 4, "Petrol")

service.add_vehicle(bike1)
service.add_vehicle(car1)
service.add_vehicle(car2)

# Customer
cus1 = Customer("C1", "Ali")
service.register_customer(cus1)


# Menu
while True:
    print("\n--- Quick Ride Rental ---")
    print("1. Show Vehicles")
    print("2. Rent Vehicle")
    print("3. Return Vehicle")
    print("4. View My Vehicle")
    print("5. Exit")
    
    choice = input("Enter your choice: ")
    
    if choice == "1":
        service.show_available_vehicles()
        
    elif choice == "2":
        vid = input("Enter vehicle ID: ")
        days = int(input("Enter number of days: "))
        service.rent_vehicle("C1", vid, days)
        
    elif choice == "3":
        service.return_vehicle("C1")
        
    elif choice == "4":
        cus1.view_rented_vehicle()
        
    elif choice == "5":
        print("Exiting...")
        break
        
    else:
        print("Invalid choice")
```
