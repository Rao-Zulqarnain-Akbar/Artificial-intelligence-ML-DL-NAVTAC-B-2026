```python
capital = {'Pakistan':'Islamabad','India':'Delhi','Canada':'Ottowa'}
del capital['India']
print(capital)
```

```python
capital = {'Pakistan':'Islamabad','India':'Delhi','Canada':'Ottowa'}
capital.clear()
print(capital)
```

```python
capital = {'Pakistan':'Islamabad','India':'Delhi','Canada':'Ottowa'}
capital['India'] = 'New Dehli'
print(capital)
```

```python
country_capital = {'Pakistan':'Islamabad','India':'Delhi','Canada':'Ottowa'}
for i in country_capital:
    print(i)
print()    
for i in country_capital:
    capital = country_capital[i]
    print(capital)
```
