# List iterator

```python
a = [1,2,3,4]
iterator = iter(a)
print(next(iterator))
print(next(iterator))
print(next(iterator))
print(next(iterator))
```

# Set iterator

```python
a = {'Ali',19,'Corvit'}
iterator = iter(a)
print(next(iterator))
print(next(iterator))
print(next(iterator))
```

# Tuple iterator

```python
num = (0,4,6,3)
iterator = iter(num)
print(next(iterator))
print(next(iterator))
```

# Dictonary iterator

```python
detail ={"Name" : "Ali", "Age" : 20, "Class" : "Corvit"}
it = iter(detail.items())
print(next(it))
print(next(it))
print(next(it))
```

```python
c = """This is
a multi-line
string"""
it = iter(c)
print(next(it))
print(next(it))
print(next(it))
print(next(it))
```

# Lambda

```python
sqr = lambda x : x**2
sqr(5)
```

# Lambda using loop

```python
check = lambda x : 'Even' if x%2==0 else 'odd'
check(5)
```

# Map function

```python
sqr = lambda x : x**2
a = [2,3,4,5,6,7,8]
sqr.val = map(sqr,a)
print(list(sqr.val))
```

# Filter function

```python
def iseven(x):
    return x % 2 == 0
even_nums = filter(iseven,a)
print(list(even_nums))    
```
