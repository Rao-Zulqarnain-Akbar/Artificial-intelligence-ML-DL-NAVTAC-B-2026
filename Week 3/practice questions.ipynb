# Revers a list

```python
list = [8,77,99,33,22,44]
for i in list[::-1]:
    print(i)
```

# Add and average

```python
a = [8, 77, 99, 33, 22, 44]

total = sum(a)
print("Sum:", total)
average = total / len(a)
print("Average:", average)
```

# customer data

```python
customer_orders = ['Mango','Dates','Apple','PineApple','Apple','Dates','Mango','Dates','Apple']
unique_values = set(customer_orders)
print(unique_values)
for item in unique_values:
    count = 0
    for order in customer_orders:
        if item == order:
            count += 1
    print(item,':',count)
    
```

# print even value

```python
student_Id = [11,22,33,11,22,33,44,66,66]
for n in student_Id:
    if n % 2 == 0:
        print(n)
```
