## Label Encoding

```python
shirt_size = ['small', 'medium', 'large', 'medium', "small", "large", "small", "large", "medium"]
shirt_color = ["white", "black", "grey", "white", "black", "grey", "black", "black", "white"]

#price = [10, ]

# basket = ['apple', 'orange', 'grape', 'strawberry', 'melon', 'plum', 'banana', 'melon', 'plum', 'plum', 'grape', 'watermelon', 'melon', 'orange']
```

```python

shirt_size = ['small', 'medium', 'large', 'medium', "small", "large", "small", "large", "medium"]

# Check if it is a nominal or ordinal (where order exist) feature? >> Ordinal 
# Which Encoding technique should be used? >> order exists, So applying >> Label Encoder

#  One-Hot Encoder (Where no order exist) 
```

```python
# Ecoding scheme
small = 1
medium = 2
large = 3

encoded_values = []

for i in shirt_size:
    print(f"Shirt Size: ", i)
    if i == 'small':
        encoded_values.append(small)
        print(f"Encoded Value: ", small)
    elif i == 'medium':
        encoded_values.append(medium)
        print(f"Encoded Value: ", medium)
    else:
        encoded_values.append(large)
        print(f"Encoded Value: ", large)
        

        
        
    
```

```python
print(shirt_size)
```

```python
print(encoded_values)
```

```python
from sklearn.preprocessing import LabelEncoder # import encoder

encoder = LabelEncoder() # intialize encoder

labels = encoder.fit_transform(shirt_size)
print(labels) 
```

```python
encoder.classes_
```

```python
for i in range(len(encoder.classes_)):
    print(f"{i} : {encoder.classes_[i]}")
```

```python
# convert the numerical labels back to the original categorical
encoder.inverse_transform(labels)
```

```python
labels = encoder.fit_transform(labels).reshape(-1, 1)
labels
```

**LabelEncoder does not support custom mappings directly** — it's designed to automatically assign integer labels in alphabetical order of the classes it sees.

However, you can simulate giving your own mapping in a few different ways depending on your needs:

```python

shirt_size = ["L", "M", 'S', 'M', 'L', 'S', 'L', 'M']   #array
custom_mapping = {'S': 1, 'M': 2, 'L': 3}   #dictonary


import pandas as pd
encoded = pd.Series(shirt_size).map(custom_mapping)
print(encoded.tolist()) # Convert it to list 
```

```python
shirt_size = ['small', 'medium', 'large', 'medium', "small", "large", "small", "large", "medium"]
custom_mapping = {'small': 0, 'medium': 1, 'large': 2}
encoded = pd.Series(shirt_size).map(custom_mapping)
print(encoded.tolist()) 
```

```python
# Reverse Mapping
reverse_mapping = {v: k for k, v in custom_mapping.items()}
decoded = encoded.map(reverse_mapping)
print(decoded.tolist())
```

## One Hot Encoding

OneHotEncoder will represent each color as a vector of binary values. 

`shirt_color = ["white", "black", "grey", "white", "black", "grey", "black", "black", "white"]`

Each color will be converted into a vector of 0s and 1s, with the 1 in the position of the color in the encoding.

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import numpy as np

#basket = ['apple', 'orange', 'grape', 'strawberry', 'melon', 'plum', 'banana', 'melon', 'plum', 'plum', 'grape', 'watermelon', 'melon', 'orange']

shirt_color = ["white", "black", "grey", "white", "black", "grey", "black", "black", "white"] # 1d array

'''
OneHotEncoder expects a 2D array as input, where each row is a separate sample and each column represents a feature. 
Even if your data is a 1D array, like the shirt_color list, it needs to be reshaped to fit the expected input shape 
for scikit-learn's OneHotEncoder.
'''
shirt_color = np.array(shirt_color).reshape(-1, 1) # convert into 2d array 

# -1: Means "calculate the number of rows automatically"
# 1: Means "there should be 1 column".

onehot_encoder = OneHotEncoder()
onehot_labels = onehot_encoder.fit_transform(shirt_color).toarray()
onehot_labels
```

```python
shirt_color
```

```python
onehot_labels
```

```python
# Check the mappings
print("Categories (Mappings):", onehot_encoder.categories_)
```

### Applying OneHot Encoder on dataframe 

```python
shirt_color = ["white", "black", "grey", "white", "black", "grey", "black", "black", "white"] # 1d array

df = pd.DataFrame({"shirt_color":shirt_color})

df
```

```python
onehot_encoder = OneHotEncoder()

# Reshape the data (because OneHotEncoder expects 2D input)
shirt_color_reshaped = df['shirt_color'].values.reshape(-1, 1)
onehot_labels = onehot_encoder.fit_transform(df['shirt_color'].values.reshape(-1, 1))

# Fit and transform the data
encoded_colors = onehot_encoder.fit_transform(shirt_color_reshaped)
encoded_colors.toarray()
```

```python
# Or do it as:
onehot_labels = onehot_encoder.fit_transform(df['shirt_color'].values.reshape(-1, 1))
onehot_labels.toarray()
```

```python
# Check the mappings
print("Categories (Mappings):", onehot_encoder.categories_)
```

####  Using pandas.get_dummies()  `More Easier way of One-Hot Encoding`

- Pandas provides pd.get_dummies() function which takes a dataframe and returns a one-hot encoded dataframe right away.

```python

shirt_color = ["white", "black", "grey", "white", "black", "grey", "black", "black", "white"]
import pandas as pd

df = pd.DataFrame(shirt_color, columns = ['Color'])
df
```

```python
df_encoded = pd.get_dummies(df) # apply one-hot encoder
print(df_encoded)
```

```python
df_encoded = pd.get_dummies(df,  dtype=int) # apply one-hot encoder
print(df_encoded)
```

# 2. Feature Scaling

```python
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df
```

```python
iris_df.describe()
```

## StandardScaler()

StandardScaler() in Scikit-Learn scales the values so that their mean is 0 and variance is 1 — Gaussian distribution. 

It is very important to convert dataset so that it follows the Gaussian distribution in some ML algorithms. 

For example, linear regression or logistic regression models assume that data follow the Gaussian distribution.


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
standard_iris = scaler.fit_transform(iris_df)
standard_iris = pd.DataFrame(standard_iris, columns = iris.feature_names)
standard_iris
```

```python
standard_iris.describe()
```

```python
standard_iris.mean()
```

```python
standard_iris.var()
```

##  MinMaxScaler()

converts data into some value in the range [0, 1], or else in the rnage [-1, 1] if there are negative values. 

With the iris data, since there are no negative values, the range of data should be between 0 and 1.

```python
from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler()
minmax_iris = minmax.fit_transform(iris_df)
minmax_iris = pd.DataFrame(minmax_iris, columns = iris.feature_names)
minmax_iris
```

```python
minmax_iris.describe()
```

```python
minmax_iris.min()
```

```python
minmax_iris.max()
```
