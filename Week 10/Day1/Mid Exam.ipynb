```python

area = [500, 800, 1000, 1200, 1500, 1800, 2000]
price = [100, 150, 180, 210, 260, 300, 330]

best_w = 0
best_b = 0
lowest_mse = float('inf')


w = 0.10
while w <= 0.20:
    b = 0
    while b <= 35:
        total_error = 0
        
        # Calculate MSE
        for i in range(len(area)):
            predicted = w * area[i] + b
            error = predicted - price[i]
            total_error += error ** 2
        
        mse = total_error / len(area)
        
        # Check lowest MSE
        if mse < lowest_mse:
            lowest_mse = mse
            best_w = round(w, 2)
            best_b = b
        
        b += 5
    w += 0.01

# Best values
print("Best w =", best_w)
print("Best b =", best_b)
print("Lowest MSE =", round(lowest_mse, 4))

# Final predictions
print("\nCompare Actual vs Predicted")
for i in range(len(area)):
    pred = best_w * area[i] + best_b
    print("Area:", area[i], "Actual:", price[i], "Predicted:", pred)

# Test new values
test_area = [700, 900, 1300, 1700, 2100]

print("\nNew Predictions")
for x in test_area:
    pred = best_w * x + best_b
    print("Area:", x, "Predicted Price:", pred)
```
