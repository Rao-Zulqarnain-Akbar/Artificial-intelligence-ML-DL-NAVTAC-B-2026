```python
# vector = 1d
# matrix = 2d
# Tensor = nd
vector = np.array([1, 2, 3])
print("Vector: ", vector)

matrix = np.array([[1, 2, 3], 
                   [4, 5, 6]])
print("Matrix: ", matrix)

tensor = np.array([[[1, 2, 44], [3, 4, 55]], 
                   [[5, 6, 66], [7, 8, 88]]])
print("Tensor: ", tensor)
```

```python
arr = np.arange(12)
print("Original array ", arr, "Shape: ", arr.shape)

reshaped = arr.reshape((3, 4))
print("\n Reshaped array ", reshaped, "Shape: ", reshaped.shape)

flattened = reshaped.flatten() # Convert multi-dimensional array to 1D
print("\n Flattened array ", flattened, )

# ravel (returns view, instead of copy)
raveled = reshaped.ravel()
print("\n raveled array ", raveled)

# Transpose
transpose = reshaped.T
print("\n Transposed array ", transpose)
```

```python
a = [[10, 20, 30], [40, 50, 60]]  # List

print(a[0][1])  # 20
print(a[1])     # [40 50 60]
print(a[:][1])  # column 1: [20 50]
print(a[0:2][1:])  # rows 0-1, cols 1-2
```

```python
a = np.array([[10, 20, 30], [40, 50, 60]])

print(a[0, 1])  # 20
print(a[1])     # [40 50 60]
print(a[:, 1])  # column 1: [20 50]
print(a[0:2, 1:])  # rows 0-1, cols 1-2
```
