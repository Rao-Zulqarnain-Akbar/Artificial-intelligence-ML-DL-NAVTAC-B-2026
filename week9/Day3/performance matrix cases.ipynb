# Case1 w=0 ,b=1

```python
area = [0, 1, 2, 3, 4, 5]
price = [0, 2, 4, 6, 8, 10]

# f_x = w * x + b

w = 0
b = 1
samples = len(area)

x = area
y = price  

total_sse = 0

for i in range(samples):
    f_x = w * x[i] + b
    # print(f"x = {x[i]}, f_x: {f_x}")
    error = y[i] - f_x
    sse = error ** 2
    total_sse += sse
    
    print(f"Actual price y = {y[i]}, Predicted value: {f_x}, Error: {error} SSE: {sse}")

print(f"SSE: {sse}")
MSE = total_sse/samples
print(f"MSE: {MSE}")
```
